```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
## STEP 1: Load dataset
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
## STEP 2: Data Preprocessing
## Drop unnecessary columns
data  = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
data
```

```python
## step 3: Catagorize varaibles
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data

```

```python
## step 4: One Hot Encoding for 'Geography' column
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(), columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
## combine all One hot encoded columns with original data
data = pd.concat([data.drop('Geography',axis=1), geo_encoded_df],axis=1)
data.head()
```

```python
## Step 5: Save the encode and scalars - transformation
with open('label_encoder_gender.pk1','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pk1','wb') as file:
    pickle.dump(onehot_encoder_geo,file)

```

```python
## step 6: Divide the dataset into independent and depedent feature
x = data.drop('Exited',axis=1) ## dependent
y =data['Exited'] ## independent
```

```python
## step 7: Split the data into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
## step 8: Scale the features
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
## Step 9: Save the data in pickle
with open('scalar.pk1', 'wb') as file:
    pickle.dump(scalar,file)
```

```python
x_train.shape[1]
```

```python

## step 10: -- ANN implmentation

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
## step 11: Build the ANN model
model = Sequential(
    [
        Dense(64,activation='relu', input_shape=(x_train.shape[1],)), ##HL1 connected with input layer
        Dense(32, activation='relu'), ##HL2
        Dense(1,activation='sigmoid') ##output layer
    ]
)
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
## step 12: compile the model for backward propogation
model.compile(optimizer=opt,loss="binary_crossentropy", metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

## Step 13: Setup Tensorboard
log_dir = "logs/fit"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callblack=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## step 13: Setup early stoppings
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
## step 14: Training the model
history = model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callblack, early_stopping_callback]
)
```

```python
## step 15: Save the model
model.save('model.h5')
```
